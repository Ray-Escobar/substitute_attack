```python
import numpy as np
import matplotlib.pyplot as plt


def create_attack_table(title, y_goog, y_res, y_dense):


    plt.title(title)
    plt.xlabel("Data Distributions")
    plt.ylabel("Accuracy")


    plt.ylim([0, 100])
    plt.yticks(np.arange(0, 110, 10))

    x = np.array(['Symmetric', 'Cross-Section', 'Disjoint'])

    # plt.step(x, y + 2, label='pre (default)')
    plt.plot(x, y_goog, 'o--', color='blue', alpha=0.3, label='goog')

    # plt.step(x, y + 1, where='mid', label='mid')
    plt.plot(x, y_res, 'o--', color='red',   alpha=0.3, label='res')

    # plt.step(x, y, where='post', label='post')
    plt.plot(x, y_dense, 'o--', color='green', alpha=0.3, label='dense')

    plt.grid(axis='x', color='0.95')
    plt.legend(loc="lower right")
    # plt.title('plt.step(where=...)')

    plt.show()
```

# PGD Tables

```python

title = 'PGD Adversarial Accuracy (Substitute: GoogLeNet)'

y_goog  = np.array([17, 44, 60])
y_res   = np.array([78, 82, 85])
y_dense = np.array([85, 88, 88])

create_attack_table(title, y_goog, y_res, y_dense)
```
