```python
import numpy as np
import matplotlib.pyplot as plt
   
Women = [115, 215, 250, 200]
Men = [114, 230, 510, 370]
Bi  = [112, 220, 514, 360]
  
n=4
r = np.arange(n)
width = 0.25
  
  
plt.bar(r, Women, color = 'b',
        width = width, edgecolor = 'black',
        label='Women')
plt.bar(r + width, Men, color = 'g',
        width = width, edgecolor = 'black',
        label='Men')
plt.bar(r + width + width, Bi, color = 'r',
        width = width, edgecolor = 'black',
        label='Bi')
  
plt.xlabel("Year")
plt.ylabel("Number of people voted")
plt.title("Number of people voted in each year")
  
# plt.grid(linestyle='--')
plt.xticks(r + width/2,['2018','2019','2020','2021'])
plt.legend()
  
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
  
N = 5
ind = np.arange(N) 
width = 0.25

xvals = [0.5, 0.48, 0.52, 0.3, 0.37]
bar1 = plt.bar(ind, xvals, width, color = 'r')
  
yvals = [0.52, 0.42, 0.30, 0.31, 0.32]
bar2 = plt.bar(ind+width, yvals, width, color='g')
  
zvals = [0.30, 0.37, 0.20, 0.22, 0.29]
bar3 = plt.bar(ind+width*2, zvals, width, color = 'b')
  
plt.xlabel("Models")
plt.ylabel('Avg. Transferability %')
plt.title("Average Transferability of adversary images")
  
plt.xticks(ind+width,['ResNet-152', 'ResNet-101', 'ResNet-50', 'VGG-16', 'GoogLeNet'])
plt.legend( (bar1, bar2, bar3), ('Equal', 'Cross-Section', 'Disjoint') )
plt.show()
```
